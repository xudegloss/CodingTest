## 1. 유효한 괄호

`````python
# {, [, (인 경우에는 무조건 stack에 넣어주기.
# }, ], )인 경우에는 앞의 문자 고려하기.
# 짝이 맞으면 제거하고, 그렇지 않으면 넣어주기.

s="{}[]()"
stack=[s[0]]

for idx in range(1, len(s)):
    if s[idx]=="{" or s[idx]=="[" or s[idx]=="(":
        stack.append(s[idx])
    else:
        if s[idx]=="}" and stack[-1]=="{":
            stack.pop()
        elif s[idx]=="]" and stack[-1]=="[":
            stack.pop()
        elif s[idx]==")" and stack[-1]=="(":
            stack.pop()
        else:
            stack.append(s[idx])
            
if len(stack)==0:
    print(True)
else:
    print(False)
`````

## 2. 일일 온도

`````python
# 온도를 모두 체크하기 위하여 T에 있는 온도를 모두 고려해야 한다.
# 고려하고 싶은 온도를 stack에 넣고, 그 온도보다 높은 온도가 나올 때까지 stack에 넣는다.
# 높은 온도가 나오면 넣지 않고 길이 만큼 넣어준다.
# 끝까지 돌려도 안 나오는 경우에는 0을 넣는다.

## break : 반복문을 빠져나갈 수 있게 한다! while 반복문에서 빠진 다음에 idx에 1을 추가하면 그 다음 부분을 체크할 수 있다.

T=[73, 74, 75, 71, 69, 72, 76, 73]
result=[]

idx=0
while idx<len(T):
    stack=[T[idx]]
    for j in range(idx+1, len(T)):
        if T[j]>stack[0]:
            result.append(len(stack))
            break
        else:
            stack.append(T[j])
            if j==len(T)-1:
                result.append(0)
    idx+=1

## 맨 마지막은 무조건 0이 들어간다.
result.append(0)
print(result)
`````

`````python
## 다시 풀어보기!

T=[73, 74, 75, 71, 69, 72, 76, 73]
result=[]

idx=0
while idx<len(T):
    stack=[T[idx]]
    for i in range(idx+1, len(T)):
        if stack[0]<T[i]:
            result.append(len(stack))
            break
        else:
            stack.append(T[i])
            if i==len(T)-1:
                result.append(0)
    idx+=1

result.append(0)
print(result)
`````

## 문제1

연속적으로 두번 중복된 문자 제거 문제

주어진 입력 s에 대하여, 연속적으로 같은 문자가 오는 경우에는 연속된 두 문자들을 문자열에서 제거하시오. 최종적으로 러턴할 문자열에는 연속된 문자들이 없어야 한다.

예) "abbac" 가 주어진 경우 "bb"가 연속된문자열이기 때문에 제거하면 "aac"가 된고, 다시 "aa"가 연속된 문자열이기 때문에 제거하여 최종적으로 "c" 문자열만 남게된다.

주의) O(N) 시간에 풀리도록 하시오.

***결과***
```
6
```


`````python
# 맨 처음 1개는 넣어둔다.
# 문자를 넣을 때, 앞의 문자와 동일한지 체크한다.
# 동일한 경우에는 제거하고, 동일하지 않은 경우에는 stack에 넣어준다.

def remove_dups(s):
    stack=[s[0]]
    for idx in range(1, len(s)):
        # 길이가 0인 경우에는 그냥 넣어주기.
        if len(stack)==0:
            stack.append(s[idx])
        # 앞 뒤가 같으면 제거하기.
        elif s[idx]==stack[-1]:
            stack.pop()
        # 다른 경우는 그냥 넣어주기.
        else:
            stack.append(s[idx])
    return "".join(map(str, stack))
    pass
    

print(remove_dups("caaa"))
print(remove_dups("accab"))
print(remove_dups("baccaaac"))
`````

## 문제3

프린터 문제

일반적인 프린터는 인쇄 요청이 들어온 순서대로 인쇄합니다. 그렇기 때문에 중요한 문서가 나중에 인쇄될 수 있습니다. 이런 문제를 보완하기 위해 중요도가 높은 문서를 먼저 인쇄하는 프린터를 개발했습니다. 이 새롭게 개발한 프린터는 아래와 같은 방식으로 인쇄 작업을 수행합니다.

1. 인쇄 대기목록의 가장 앞에 있는 문서(J)를 대기목록에서 꺼냅니다.
2. 나머지 인쇄 대기목록에서 J보다 중요도가 높은 문서가 한 개라도 존재하면 J를 대기목록의 가장 마지막에 넣습니다.
3. 그렇지 않으면 J를 인쇄합니다.

예를 들어, 4개의 문서(A, B, C, D)가 순서대로 인쇄 대기목록에 있고 중요도가 2 1 3 2 라면 C D A B 순으로 인쇄하게 됩니다.

내가 인쇄를 요청한 문서가 몇 번째로 인쇄되는지 알고 싶습니다. 위의 예에서 C는 1번째로, A는 3번째로 인쇄됩니다.

현재 대기목록에 있는 문서의 중요도가 순서대로 담긴 배열 priorities와 내가 인쇄를 요청한 문서가 현재 대기목록의 어떤 위치에 있는지를 알려주는 location이 매개변수로 주어질 때, 내가 인쇄를 요청한 문서가 몇 번째로 인쇄되는지 return 하도록 함수를 작성해주세요.

***결과***
```
1
4
2
5
````


`````python
from collections import deque

def printer_order(priorities, location):
    queue =  deque([(val, idx) for idx, val in enumerate(priorities)])
    answer = 0
    
    while queue:
        cur=queue.popleft()
        if queue and cur[0] < max(queue)[0]:
            queue.append(cur)
        else:
            answer+=1 # 위치 정보를 의미한다.
            if cur[1]==location:
                return answer

print(printer_order([2, 1, 3, 2], 2))    
print(printer_order([2, 1, 3, 2], 1))
print(printer_order([2, 1, 3, 2], 3))
print(printer_order([1, 1, 9, 1, 1, 1], 0))
`````

`````python
## 다시 풀어보기!!!

from collections import deque

def printer_order(priorities, location):
    queue =  deque([(val, idx) for idx, val in enumerate(priorities)])
    answer = 0 # 인덱스를 의미한다. (위치 정보)
    
    while queue:
        cur=queue.popleft()
        if queue and cur[0] < max(queue)[0]:
            queue.append(cur)
        else:
            answer+=1
            if cur[1]==location:
                return answer

print(printer_order([2, 1, 3, 2], 2))    
print(printer_order([2, 1, 3, 2], 1))
print(printer_order([2, 1, 3, 2], 3))
print(printer_order([1, 1, 9, 1, 1, 1], 0))
`````
